# Literature Review

**Early Developments in Factor-Based Investing**

The foundational work by {cite}`fama_1992` identified size and book-to-market (B/M) ratios as key factors in explaining stock returns' variations. Their research demonstrated that these factors offered better explanatory power than the traditional Capital Asset Pricing Model (CAPM), which relied solely on market beta. This laid the groundwork for factor-based investing, where characteristics like size and value (high B/M ratios) became crucial in building portfolios and managing risk.

Later, {cite}`fama_1993` expanded their analysis by including bond market factors and integrating size and value factors into a comprehensive model to explain returns across different asset classes. Although this model was developed primarily for mature markets, it raised questions about its applicability in emerging markets like India, which have unique risk profiles, liquidity issues, and regulatory environments. The distinct characteristics of emerging markets often require a more tailored approach to factor investing, as traditional factors like size and value may behave differently due to regional influences.

**Factor Models and Value in Emerging Markets**

{cite}`blackburn_2020` conducted a detailed study of the B/M effect across 25 emerging markets, concluding that only the tangible component of B/M, which reflects past accounting performance, was significantly related to expected returns. Their findings suggest the importance of focusing on accounting-based factors, like profitability, rather than relying solely on market-driven price changes, which can be more prone to inefficiencies in less developed markets. This insight is especially relevant for constructing portfolios in emerging markets like India, indicating that tangible, profitability-linked factors might provide better predictions for returns.

Similarly, {cite}`asness_2013` explored the profitability factor and demonstrated its robustness in a value-oriented context. He showed that highly profitable companies tend to outperform, regardless of their size or B/M ratios. In emerging markets, where the quality of earnings can vary widely, this profitability premium can be a crucial element in building a resilient portfolio.

**Value and Profitability in India**

In the Indian context, recent research by {cite}`farhana_2024` revisited the Fama-French three-factor model over the period from 2000 to 2023. Mehta found that while both size and value factors are important, the size factor has a more significant impact on stock returns. Specifically, large-cap firms with low B/M ratios (BL portfolios) offer greater explanatory power in returns. This suggests that when constructing a value-focused portfolio in India, prioritizing larger, more liquid firms could lead to a more consistent strategy, especially when combined with profitability as a secondary criterion.

Building on this, {cite}`agarwalla_2014` introduced a four-factor model to the Indian equity market, adding momentum to enhance the explanatory power of the size, value, and market factors. Their findings revealed that value (HML) and profitability factors collectively provided substantial returns, reaffirming the relevance of these fundamentals in emerging markets. Their rigorous methodology, which accounted for survivorship bias, underscored the importance of selecting investable, liquid portfolios—a critical factor in markets where transparency is often limited.

**Profitability as a Factor**

Although the Fama-French models emphasize size and value, recent studies, particularly in emerging markets, highlight profitability as a vital determinant of stock returns. Profitability, often measured by returns on assets or equity, offers a complementary perspective to value by assessing a firm's operational efficiency and future earnings potential.

Globally, {cite}`asness_2013` found that profitability factors, like value and momentum, exhibit strong correlations across asset classes and markets. They observed that profitability and value factors are negatively correlated, suggesting that incorporating highly profitable firms into a value portfolio can help balance risk exposures, particularly in volatile markets like India. This negative correlation indicates that while value stocks may experience periods of cyclical volatility, the inclusion of profitable firms can provide a buffer, offering resilience during market downturns.

**Behavioral and Liquidity Considerations in Emerging Markets**

Behavioral finance provides an additional lens through which to understand factor premiums in emerging markets. {cite}`barberis_2003` argue that investor biases, such as overconfidence and representativeness, can lead to mispricing and return anomalies, particularly in less efficient markets. These behavioral influences can help explain the persistence of factors like value and momentum, especially in markets like India, where inefficiencies are more common.

Liquidity is another critical consideration in emerging markets. {cite}`amihud_1986` demonstrated that liquidity constraints have a significant impact on asset pricing, with illiquid assets commanding higher expected returns to compensate for transaction costs. In India, where liquidity varies significantly across firms, considering liquidity as an additional factor alongside profitability and value could strengthen portfolio construction by mitigating the risks associated with thinly traded stocks.

**Momentum and Seasonality in Indian Markets**

Momentum, first introduced by {cite}`jegadeesh_1993` and further examined by {cite}`agarwalla_2014` for Indian equities, is an influential factor in emerging markets. The persistence of momentum indicates that past winners may continue to perform well, challenging the efficient market hypothesis. {cite}`rouwenhorst_2001` found similar patterns across emerging markets, underscoring momentum as a potential focus for Indian portfolios, where market inefficiencies may lead to sustained price trends.

Additionally, {cite}`harshita_2019` identified unique seasonal patterns in the Indian market, where small firms tend to generate higher returns in November and December. This "calendar effect" could provide a tactical advantage when structuring profitability-based strategies, particularly with smaller-cap stocks that display seasonal price variations. Incorporating seasonality could thus help optimize entry and exit points for profitability-driven investments.

**Avoiding Index Rebalancing Crowds**

Managing a value and profitability-focused portfolio also requires avoiding inefficiencies related to index rebalancing. {cite}`arnott_2023` observed that discretionary deletions often outperform additions in major indices, suggesting that index rebalancing can introduce temporary price distortions. This effect likely applies to Indian indices as well, where timing trades ahead of or after index reconstitution can help capture excess returns. By focusing on fundamental metrics rather than mechanical index adjustments, investors can further optimize long-term performance.

**Conclusion**

Research on factor-based investing in emerging markets, particularly in India, highlights the crucial role of value and profitability in portfolio construction. Studies by {cite}`farhana_2024` and {cite}`agarwalla_2014` confirm that value remains a dominant factor, while profitability adds significant explanatory power. Additionally, liquidity, behavioral considerations, and unique market phenomena such as calendar effects and index rebalancing inefficiencies provide valuable insights for enhancing portfolio resilience and performance. By integrating these factors and acknowledging India’s unique market dynamics, investors can develop a robust strategy to capture long-term returns.

Incorporating momentum and seasonality can further refine investment strategies, as shown by the persistence of momentum and calendar anomalies in the Indian market. Understanding behavioral biases and liquidity constraints unique to emerging markets like India is essential for optimizing factor-based approaches. Moreover, staying mindful of index rebalancing effects can help investors avoid pitfalls associated with mechanical trading strategies.

As the Indian market continues to evolve, ongoing research and adaptation of factor models will be crucial. Investors should remain alert to structural changes, regulatory developments, and macroeconomic shifts that could influence factor performance. By maintaining a dynamic and informed approach, investors can effectively harness the benefits of factor-based investing in India’s emerging market landscape.

# Introduction

Factor-based investing has become a cornerstone of modern asset management, reshaping portfolio construction and risk management. This approach first gained prominence with the pioneering work of Fama and French in the early 1990s, whose factor models identified key characteristics—such as size and value—that better explained cross-sectional variations in stock returns compared to the traditional Capital Asset Pricing Model (CAPM), which relied solely on market beta. By emphasizing the importance of book-to-market (B/M) ratios and firm size, Fama and French laid the groundwork for multifactor models that have since become central to portfolio construction. However, while most of this research has focused on developed markets, emerging markets like India present a unique and largely untapped frontier for factor-based strategies.

This paper explores one such frontier by implementing a two-dimensional factor strategy centered on value and profitability in Indian equities. The Nifty 500—a broad index representing a significant portion of the Indian stock market—serves as the investable universe for this study. While value has long been recognized as a critical driver of returns in Indian equities, recent research suggests that profitability, especially when linked to tangible financial metrics, adds substantial explanatory power. Blackburn (2020) demonstrated that in emerging markets, profitability tied to past accounting performance, rather than market-driven price changes, serves as a more reliable predictor of future returns. This insight is particularly relevant for Indian markets, where accounting-based factors may offer an edge over purely market-driven indicators. Thus, this study integrates both value and profitability factors to construct a portfolio aimed at generating alpha in the Indian context.

The increasing interest in factor-based investing in India brings challenges unique to emerging markets. Factor models developed in the U.S. or other developed markets often face limitations when applied to markets like India, which exhibit distinct risk-return profiles and market inefficiencies. For instance, studies like Agarwalla (2014) show that while value remains a dominant factor in Indian equities, delivering robust returns, the addition of factors such as profitability or momentum can improve portfolio performance. Furthermore, emerging markets are generally more vulnerable to inefficiencies, including those from seasonal patterns and index rebalancing. Research on Indian equities, for instance, has found that smaller firms often experience significant price movements in November and December, reflecting seasonal return patterns (Harshita 2019). Such unique phenomena emphasize the need for strategies that consider these inefficiencies in portfolio construction.

In this context, this paper contributes to existing literature by addressing two often-overlooked biases in the Indian equity space: survivorship bias and look-ahead bias. Survivorship bias arises when only surviving stocks over a period are analyzed, thereby skewing results by excluding companies that were delisted or failed. Look-ahead bias, on the other hand, occurs when backtested strategies incorporate information that would not have been available at the time of trading decisions. Both biases can inflate portfolio performance expectations. This study aims to mitigate these biases by constructing a value and profitability portfolio that uses only the data available at the time of trading and includes stocks that exited the index during the analysis period.

The significance of this research extends beyond refining factor-based strategies for the Indian market; it holds broader implications for emerging markets worldwide. As globalization has expanded access to financial markets, investors increasingly seek sophisticated tools to navigate the complexities of emerging economies. Factor-based strategies incorporating accounting-driven metrics like profitability offer investors a structured method to capture returns in these markets. Additionally, by addressing common biases and incorporating insights from recent research, this paper strives to present a more accurate and implementable approach to factor-based investing in Indian equities.

In conclusion, this paper examines the effectiveness of a two-dimensional factor strategy that combines value and profitability within the Nifty 500, leveraging tools such as Bloomberg Terminal and CMIE's ProwessDX to construct a robust dataset. The ultimate aim is to generate alpha through a systematic, data-driven approach that addresses the unique challenges of investing in Indian equities. Through this research, we aim to provide a roadmap for both domestic and international investors seeking to capitalize on opportunities in India’s dynamic and evolving equity markets.

# Methodology

This study evaluates the performance of value and profitability factors within the Nifty 500 index over a ten-year period (September 2014 to September 2024). Using data from Bloomberg Terminal and CMIE’s ProwessDX, the methodology includes data collection, restructuring, factor calculations, and portfolio construction.

Calendar earnings data for each stock was collected using Bloomberg's ERN function, covering the entire 10-year period. Additional data on value metrics, profitability metrics, price data, adjustment factors, dividends, and stock identity information were also extracted. Each stock was then manually mapped to its sector using the GICS methodology on Bloomberg, ensuring consistent sector classification across data sets. These collected datasets were consolidated into a structured data frame, facilitating subsequent adjustments and analysis.

To ensure accurate price representation, dividend and stock split adjustments were applied. A custom function adjusted prices for dividends by proportionally modifying prices before the ex-dividend date to align with dividend payouts. For stock splits, price adjustments were applied proportionally based on split factors, affecting only data preceding each split date. The final adjusted prices, which accounted for both dividends and splits, provided a consistent price series for factor analysis.

The standardized value and profitability factors were calculated on a sector basis. Two sector-based z-score data frames were created, one for profitability factors and another for value factors, allowing comparability within each sector and helping isolate relative value and profitability.

A composite score was then calculated for each stock by averaging its sector-based profitability and value z-scores. This data frame also included sector mapping, earnings release dates, and the corresponding financial quarter, enabling quarter-specific analysis. 

Finally, a quarterly rebalanced portfolio was simulated, trading on the last day of each quarter. Stocks were ranked by their composite z-scores, with the top 30 selections forming the portfolio each quarter. The portfolio's performance was assessed by calculating cumulative returns over the ten-year period. For comparison, an equal-weighted Nifty 500 index, normalized to start at 1, served as a benchmark.


# Results

This section presents the findings from the analysis of a value and profitability-based investment strategy applied to stocks within the Nifty 500 index over a ten-year period (September 2014 to September 2024). The primary research question is whether a portfolio constructed based on value and profitability factors can outperform the broader Nifty 500 index.

This plot shows each stock’s position on a 2D plane, with the x-axis representing the Value Z-score and the y-axis representing the Profitability Z-score. Stocks with lower Value Z-scores and higher Profitability Z-scores are generally considered more attractive under value investing principles, as they represent undervalued yet profitable companies.

Each point on the timelapse represents a stock at a specific time, with colors distinguishing between different ranges or classifications. Over time, the movement of points reflects changes in stocks' relative value and profitability within their sectors.

**Sample Snapshot of Factor Evolution** captures the visualization at a single point in time (October 31, 2016). In this snapshot, most stocks are clustered around the center, indicating relatively neutral Z-scores for both value and profitability. A few outliers can be seen in the lower-left and upper-right quadrants. Stocks in the lower-left (with negative Z-scores on both axes) are undervalued but less profitable, while those in the upper-right (with positive Z-scores on both axes) are both relatively high in value and profitability. The color coding helps track these changes over time, providing a dynamic view of factor shifts within the Nifty 500.

The factor-based portfolio was constructed by selecting the top 30 stocks with the highest combined scores in value (P/B, P/E ratios) and profitability (ROA, ROE), recalculated on a quarterly basis. This quarterly rebalancing allowed the portfolio to remain aligned with the most attractive stocks according to the chosen factors. Cumulative returns for the portfolio were then compared with those of the Nifty 500 index, with both the portfolio and the index normalized to start at a value of 1 for a straightforward comparison over the study period. The **Value & Profitability vs Nifty 500 Benchmark** plot shows that the factor-based portfolio achieved consistently higher cumulative returns than the benchmark, especially during volatile market periods. This highlights the effectiveness of a combined value and profitability strategy in the Indian market.

Risk-adjusted performance was also evaluated using the Sharpe ratio, which revealed that the factor-based portfolio outperformed the Nifty 500 in terms of risk-adjusted returns. This superior Sharpe ratio indicates that the portfolio not only generated higher returns but also maintained a more efficient risk profile, underscoring the potential of a value-profitability approach for managing risk while maximizing returns. Additionally, quarterly rebalancing was instrumental in adapting the portfolio to changing market dynamics, as it allowed the inclusion of stocks that consistently ranked high in value and profitability. This rebalancing approach proved particularly effective during market downturns, as it frequently rotated into fundamentally strong stocks. The portfolio’s ability to mitigate risk while sustaining high returns throughout the study period highlights the practical benefits of regular rebalancing in a long-term factor-based strategy. All in all, the results reinforce the potential for a strategy focused on undervalued yet profitable stocks to generate higher returns and improve resilience across various market conditions.

## Profitability Metrics

#### 1. Gross Profit Margin
$$
\text{Gross Profit Margin} = \frac{\text{Gross Sales} - \text{Total Expenses}}{\text{Gross Sales}} \times 100
$$

#### 2. Operating Profit Margin (EBIT Margin)
$$
\text{Operating Profit Margin} = \frac{\text{Income from Operations (EBIT)}}{\text{Net Sales}} \times 100
$$

#### 3. Net Profit Margin
$$
\text{Net Profit Margin} = \frac{\text{Profit After Tax (PAT)}}{\text{Net Sales}} \times 100
$$

#### 4. Return on Assets (ROA)
$$
\text{Return on Assets (ROA)} = \frac{\text{Profit After Tax (PAT)}}{\text{Shareholders' Funds} + \text{Total Expenses}} \times 100
$$

#### 5. Return on Equity (ROE)
$$
\text{Return on Equity (ROE)} = \frac{\text{Profit After Tax (PAT)}}{\text{Shareholders' Funds}} \times 100
$$

#### 6. EBITDA Margin
$$
\text{EBITDA Margin} = \frac{\text{Income from Operations (EBIT)} + \text{Depreciation}}{\text{Net Sales}} \times 100
$$

#### 7. Return on Sales (ROS)
$$
\text{Return on Sales (ROS)} = \frac{\text{Profit After Tax (PAT)}}{\text{Net Sales}} \times 100
$$

#### 8. Cash Flow to Sales Ratio
$$
\text{Cash Flow to Sales Ratio} = \frac{\text{Cash Flow from Operating Activities}}{\text{Net Sales}} \times 100
$$

#### 9. Earnings Per Share (EPS Diluted)
$$
\text{Earnings Per Share (EPS Diluted)} = \text{Diluted EPS (as reported)}
$$

#### 10. Operating Expenses to Sales Ratio
$$
\text{Operating Expenses to Sales Ratio} = \frac{\text{Total Expenses}}{\text{Net Sales}} \times 100
$$

#### 11. Interest Coverage Ratio
$$
\text{Interest Coverage Ratio} = \frac{\text{Income from Operations (EBIT)}}{\text{Interest Expenses}}
$$
